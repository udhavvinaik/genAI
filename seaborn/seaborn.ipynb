#Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x_v = np.array([1,2,3,4,5,6,7,10,20])
y_v = x_v**2

df = pd.DataFrame({"x_values":x_v,"y_values":y_v})

#x and y represent coloumn names in out data frame
sns.lineplot(x="x_values",y="y_values",data = df)
```

```python
#loading a dataset 
d = sns.load_dataset("penguins")
d

sns.lineplot(x="bill_length_mm",y="flipper_length_mm",hue="island",palette="viridis",markers=["o",">","*"],data=d,dashes=True)
plt.grid()
```

barplot

```python
d = sns.load_dataset("penguins")
print(d.tail())
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
order_2 = ["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=d,palette="summer",hue="sex",order=order_1,hue_order=order_2)
```

```python
d=d.head(30)
sns.barplot(x="flipper_length_mm",y="bill_length_mm",data=d,palette="summer",orient="h")
plt.grid()
```

HISTOGRAM

```python

sns.displot(d["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True)
```

SCATTERPLOT

```python
sns.scatterplot(x="flipper_length_mm",y="bill_length_mm",data=d,hue='island',style='island',size='island')
```
