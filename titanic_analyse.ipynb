```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = sns.load_dataset("titanic")
```

```python
d.describe()
```

```python
d.head()
```

survival count by gender

```python
sns.countplot(data=d,x="survived",hue="sex")
```

 Survival Count by Passenger Class

```python
sns.countplot(data=d,x="class",hue="survived")
```

Age Distribution

```python
sns.histplot(data=d.dropna(),x="age",kde=True,bins=30)
plt.title("Age Distribution")
```

Age vs Class Box Plot

```python
sns.boxplot(data = d,x="class",y="age",hue="sex")
```

Heatmap of Correlations

```python
sns.heatmap(data=d[["survived","pclass","age","sibsp","parch","fare"]].corr(),cmap="coolwarm",annot=True)
```

Fare Distribution by Class & Survival

```python
sns.barplot(data=d,x="class",y="fare",hue="survived",palette=["cyan","grey"])
```

```python
sns.pairplot(d[['survived', 'class', 'age', 'fare']].dropna(), hue='survived')
```
