```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["director"].fillna("Unknown",inplace=True)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

MOVIES VS TV SHOWS

```python
sns.countplot(data=df,x="type")
plt.title("movies vs tv shows")
```

```python
release_df = df["release_year"].value_counts().sort_index().reset_index()
release_df
sns.lineplot(data=release_df,x="release_year",y="count")
plt.title("no of movies per year")
plt.grid()
```

Top Countries Producing Content

```python
country_df = df["country"].value_counts()
country_df = country_df.head(10).reset_index()
country_df
sns.barplot(data=country_df,x="country",y="count",palette="coolwarm",hue="country")
plt.xticks(rotation=45)
plt.title("Top Countries Producing Content")
```

```python
from collections import Counter
genre_series = df["listed_in"].dropna().str.split(", ")
all_genres = [genre for sublist in genre_series for genre in sublist]

genre_counts = Counter(all_genres)
top_genres = pd.DataFrame(genre_counts.most_common(10), columns=["genre", "count"])

sns.barplot(data=top_genres, x="genre", y="count", palette="viridis",hue="genre")
plt.title("Top 10 Genres")
plt.xticks(rotation=90)
```

```python
df.head()
```

```python
# Step 1: Split the duration column
df[["duration_int", "duration_type"]] = df["duration"].str.split(" ", expand=True)

# Step 2: Convert duration_int to numeric
df["duration_int"] = df["duration_int"].astype(int)

df.head()
```

```python
movie_df = df[df["duration_type"]=="min"]
sns.histplot(data=movie_df,x="duration_int")
plt.title("Movies screen time",fontsize=40,color="blue")
```

```python
shows_df = df[df["duration_type"]=="Seasons"]
sns.countplot(data=shows_df,x="duration_int")
plt.title("no of seasons for shows",fontsize=30,color="blue")
```

```python
sns.countplot(data=df,x="rating",hue="type")
plt.xticks(rotation=90)
plt.title("Rating Distribution",fontsize=30)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Prepare the cast data as you've done
cast_df = df["cast"].str.split(", ")
cast_df = [cast for sublist in cast_df for cast in sublist]
cast_counts = Counter(cast_df)
cast_df = pd.DataFrame(cast_counts.most_common(50), columns=["cast", "count"])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(dict(cast_df.values))

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.title("Top 10 Cast Members in Movies/TV Shows", fontsize=16)
plt.show()
```
